```python
import boto3
import pytesseract
import pdf2image
from pdf2image import convert_from_path
import cv2


LANG = "en"
AWS_REGION = 'ap-south-1'
translate = boto3.client(service_name='translate',
                         region_name=AWS_REGION,
                         use_ssl=True)

class pdf_processor:
    def __init__(self,image_path):
        self.image_path=image_path
        self.image=cv2.imread(str(self.image_path))
        self.grey_img=cv2.cvtColor(self.image,cv2.COLOR_BGR2GRAY)
        self.thresholding_image=cv2.threshold(self.grey_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
        self.bitwise_image=cv2.bitwise_not(self.thresholding_image)

    def find_contours(self):
        self.img_contours, self.img_hierarchy = cv2.findContours(self.bitwise_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    def filter_contours_and_leave_only_rectangles(self):
        self.rectangular_contours = []
        for contour in self.img_contours:
            peri = cv2.arcLength(contour, True)
            approx = cv2.approxPolyDP(contour, 0.02 * peri, True)
            if len(approx) == 4:
                self.rectangular_contours.append(approx)
        self.rectangular_contours=self.rectangular_contours[::-1]
        #return self.rectangular_contours
    
    def crop_rectangular_contours(self):
        count=1
        for i in range(0,len(self.rectangular_contours)):
            x,y,w,h=cv2.boundingRect(self.rectangular_contours[i])
            cropped_image=self.thresholding_image[y:y+h,x:x+w]
            if not os.path.exists('./data/'+str(self.image_path)+'/'):
                os.makedirs('./data/'+str(self.image_path)+'/')
            path='./data/'+str(self.image_path)+'/'+'croppedimage_'+str(count)+'.jpeg'
            #print("area of contour is: ",cv2.contourArea(self.rectangular_contours[i]))
            if cv2.contourArea(self.rectangular_contours[i])>=1000:
                #print("image is : ",path)
                cv2.imwrite(path,cropped_image)
                count=count+1

    def extract_text_from_image(self,image):
        text = pytesseract.image_to_string(image,lang='mar',config='--psm 4')
        return text
    
    def converting_image_to_text(self,threshold=30):
        self.dct={}
        count=1
        total_files=len(os.listdir('./data/'+str(self.image_path)+'/'))
        print('Total files : ',total_files)
        for file in sorted(os.listdir('./data/'+str(self.image_path)+'/'),key=lambda x:int(x.split("_")[1].split(".")[0])):
            # Step 2: Preprocess the image (deskew)
            #preprocessed_image = deskew(np.array(page))

            # Step 3: Extract text using OCR
            #print("file opened:{}".format(file))
            image4=cv2.imread('./data/'+str(self.image_path)+'/'+file)
            text = self.extract_text_from_image(image4)
            try:
                result = translate.translate_text(Text=text,
                                            SourceLanguageCode="mr",
                                            TargetLanguageCode=LANG)
            except:
                print("No Text found.")
            translation = result.get('TranslatedText')
            #print('translated_text:{}'.format(translation))
            if count%2!=0:
                key=translation.strip()
                self.dct[key]=''
            else:
                self.dct[key]=translation.strip()
            count=count+1
            if min(total_files,threshold)!=threshold:
                threshold=total_files
            if count>=threshold:
                break
        return self.dct
    
os.chdir('/Users/sbhosale1/Documents/propreturns_takehome_assignment')
```

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
#from cStringIO import StringIO

def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    codec = 'utf-8'
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, codec=codec, laparams=laparams)

    with open(path, 'rb') as fp:
        interpreter = PDFPageInterpreter(rsrcmgr, device)
        password = ""
        caching = True
        pagenos = set()

        for page in PDFPage.get_pages(fp, pagenos, password=password,caching=caching, check_extractable=True):
            interpreter.process_page(page)

        text = retstr.getvalue()

    device.close()
    retstr.close()
    return text

convert_pdf_to_txt("./propertydetails.pdf")
```
